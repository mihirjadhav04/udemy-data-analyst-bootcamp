```python
#Handling missing values
# df.isnull() # identify with boolean value.
# df.isnull().any() # to indicate which columns have missing values
# df.isnull().sum() # to identify how many missing values are in each column


filled_df = df.fillna(0) # to replace missing values with 0 - sample

```

```python
filled_df
```

```python
# Fill missing values with the mean of the column
df['Sales_fillNA']=df['Sales'].fillna(df['Sales'].mean())
df
```

```python
df.rename(columns={"Date" : "Sales Date"}) # rename columns
```

```python
#change datatypes

df["Value_new"]=df["Value"].fillna(df["Value"].mean()).astype(int)
df.head()
```

```python
df["New Value"] = df["Value"].apply(lambda x: x*5) # apply function to a column
df.head(5)
```

```python
# Group by and aggregate
grouped_mean = df.groupby("Product")["Value"].mean()
print(grouped_mean)
```

```python
df
```

```python
grouped_sum = df.groupby(["Product","Region"])["Value"].sum()
print(grouped_sum)
```

```python
grouped_agg = df.groupby("Region")["Value"].agg(["mean","sum","count"])
grouped_agg
```

```python
# Merging and Joining
df1 = pd.DataFrame({"KEY" : ["A", "B", "C", "D"], "VALUE1" : [10, 20, 30, 40]})
df2 = pd.DataFrame({"KEY" : ["A", "B", "D", "E"], "VALUE2" : [50, 60, 70, 80]})   

                    
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="KEY",how="inner")
```

```python
pd.merge(df1,df2,on="KEY",how="outer")
```

```python
pd.merge(df1,df2,on="KEY",how="left")
```

```python
pd.merge(df1,df2,on="KEY",how="right")
```
